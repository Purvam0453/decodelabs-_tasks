## Step 1 - Import Libraries

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## Step 2 - Load the Dataset

```python
df = pd.read_excel('Dataset for Data Analytics (1).xlsx')
print('Shape:', df.shape)
df.head(10)
```

## Step 3 - Explore the Data

```python
# Check data types and basic info
df.info()
```

```python
# Check missing values in each column
print('Missing Values:')
print(df.isnull().sum())
```

```python
# Check for duplicate rows
print('Duplicate rows:', df.duplicated().sum())
print('Duplicate OrderIDs:', df['OrderID'].duplicated().sum())
```

```python
# Check unique values in categorical columns
for col in ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource', 'CouponCode']:
    print(f'{col}: {df[col].unique()}')
```

```python
# Check basic statistics
df.describe()
```

## Step 4 - Remove Duplicates

```python
before = len(df)
df.drop_duplicates(inplace=True)
after = len(df)
print(f'Rows before: {before}')
print(f'Rows after:  {after}')
print(f'Duplicates removed: {before - after}')
```

## Step 5 - Handle Missing Values

```python
# CouponCode has 309 missing values
# These rows simply had no coupon applied, so filling with 'NO_COUPON'
print('CouponCode null before:', df['CouponCode'].isnull().sum())
df['CouponCode'] = df['CouponCode'].fillna('NO_COUPON')
print('CouponCode null after: ', df['CouponCode'].isnull().sum())
```

## Step 6 - Fix Data Formats

```python
# Fix Date format - convert to YYYY-MM-DD (ISO 8601 standard)
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')
print('Sample dates after formatting:')
print(df['Date'].head(5).tolist())
```

```python
# Fix TotalPrice floating point precision errors
# Some rows had values like 769.3799999999999 due to float multiplication
precision_errors = df['TotalPrice'].apply(lambda x: len(str(x).split('.')[-1]) > 2).sum()
print(f'TotalPrice rows with precision errors: {precision_errors}')
df['TotalPrice'] = df['TotalPrice'].round(2)
df['UnitPrice']  = df['UnitPrice'].round(2)
print('Fixed: all values rounded to 2 decimal places')
```

```python
# Remove leading/trailing whitespace from all text columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
print('Whitespace removed from all text columns')
```

## Step 7 - Verify Cleaned Data

```python
print('=== FINAL VERIFICATION ===')
print(f'Total rows       : {len(df)}')
print(f'Total columns    : {len(df.columns)}')
print(f'Missing values   : {df.isnull().sum().sum()}')
print(f'Duplicate rows   : {df.duplicated().sum()}')
print(f'Duplicate IDs    : {df["OrderID"].duplicated().sum()}')
print()
print('Missing values per column:')
print(df.isnull().sum())
```

```python
# Preview final cleaned dataset
df.head(10)
```

## Step 8 - Save Cleaned Dataset

```python
df.to_excel('Cleaned_Dataset.xlsx', index=False)
print('Cleaned dataset saved as Cleaned_Dataset.xlsx')
```
