## Step 1 - Setup: Load Data into SQLite

```python
import pandas as pd
import sqlite3
import warnings
warnings.filterwarnings('ignore')

# Load the cleaned dataset
df = pd.read_excel('Cleaned_Dataset.xlsx')
df['Date'] = pd.to_datetime(df['Date'])

# Create in-memory SQLite database and load data into it
conn = sqlite3.connect(':memory:')
df.to_sql('sales', conn, index=False, if_exists='replace')

print('Database created successfully')
print('Table: sales')
print('Total rows loaded:', len(df))
```

```python
# Helper function to run SQL and return result as DataFrame
def run_query(query):
    return pd.read_sql_query(query, conn)

# Quick check - see all column names
result = run_query('PRAGMA table_info(sales)')
print('Columns in sales table:')
print(result[['name', 'type']].to_string(index=False))
```

## Step 2 - Basic SELECT Queries

```python
# View first 5 rows from the table
query = '''
SELECT *
FROM sales
LIMIT 5
'''
run_query(query)
```

```python
# Select only important columns
query = '''
SELECT OrderID, Product, Quantity, UnitPrice, TotalPrice
FROM sales
LIMIT 10
'''
run_query(query)
```

```python
# Count total number of orders in the dataset
query = '''
SELECT COUNT(*) AS Total_Orders
FROM sales
'''
run_query(query)
```

## Step 3 - WHERE Clause (Filtering Data)

```python
# Filter orders where TotalPrice is greater than 5000
query = '''
SELECT OrderID, Product, Quantity, TotalPrice
FROM sales
WHERE TotalPrice > 5000
LIMIT 10
'''
run_query(query)
```

```python
# See all unique products in the dataset
query = '''
SELECT DISTINCT Product
FROM sales
'''
run_query(query)
```

```python
# Count only Delivered orders
query = '''
SELECT COUNT(*) AS Delivered_Orders
FROM sales
WHERE OrderStatus = 'Delivered'
'''
run_query(query)
```

```python
# Cancelled orders with high value (above 3000)
query = '''
SELECT OrderID, Product, TotalPrice, OrderStatus
FROM sales
WHERE OrderStatus = 'Cancelled' AND TotalPrice > 3000
ORDER BY TotalPrice DESC
LIMIT 10
'''
run_query(query)
```

```python
# Orders where Quantity is more than 5
query = '''
SELECT OrderID, Product, Quantity, TotalPrice
FROM sales
WHERE Quantity > 5
ORDER BY Quantity DESC
LIMIT 10
'''
run_query(query)
```

## Step 4 - ORDER BY (Sorting Results)

```python
# Top 10 highest value orders
query = '''
SELECT OrderID, Product, Quantity, TotalPrice
FROM sales
ORDER BY TotalPrice DESC
LIMIT 10
'''
run_query(query)
```

```python
# 10 lowest value orders
query = '''
SELECT OrderID, Product, Quantity, TotalPrice
FROM sales
ORDER BY TotalPrice ASC
LIMIT 10
'''
run_query(query)
```

```python
# Sort by Product name (A to Z), then by TotalPrice (high to low)
query = '''
SELECT OrderID, Product, TotalPrice
FROM sales
ORDER BY Product ASC, TotalPrice DESC
LIMIT 15
'''
run_query(query)
```

## Step 5 - GROUP BY + Aggregations (COUNT, SUM, AVG)

```python
# Total number of orders per product
query = '''
SELECT Product,
       COUNT(*) AS Total_Orders
FROM sales
GROUP BY Product
ORDER BY Total_Orders DESC
'''
run_query(query)
```

```python
# Total revenue per product using SUM
query = '''
SELECT Product,
       ROUND(SUM(TotalPrice), 2) AS Total_Revenue
FROM sales
GROUP BY Product
ORDER BY Total_Revenue DESC
'''
run_query(query)
```

```python
# Average order value per product using AVG
query = '''
SELECT Product,
       ROUND(AVG(TotalPrice), 2) AS Avg_Order_Value
FROM sales
GROUP BY Product
ORDER BY Avg_Order_Value DESC
'''
run_query(query)
```

```python
# Order count grouped by OrderStatus
query = '''
SELECT OrderStatus,
       COUNT(*) AS Order_Count
FROM sales
GROUP BY OrderStatus
ORDER BY Order_Count DESC
'''
run_query(query)
```

```python
# Revenue summary by Payment Method
query = '''
SELECT PaymentMethod,
       COUNT(*) AS Total_Orders,
       ROUND(SUM(TotalPrice), 2) AS Total_Revenue,
       ROUND(AVG(TotalPrice), 2) AS Avg_Order_Value
FROM sales
GROUP BY PaymentMethod
ORDER BY Total_Revenue DESC
'''
run_query(query)
```

```python
# Orders and revenue by Referral Source
query = '''
SELECT ReferralSource,
       COUNT(*) AS Total_Orders,
       ROUND(SUM(TotalPrice), 2) AS Total_Revenue
FROM sales
GROUP BY ReferralSource
ORDER BY Total_Orders DESC
'''
run_query(query)
```

## Step 6 - Combined Queries (Business Questions)

```python
# Q1: Which product generates most revenue from Delivered orders only?
query = '''
SELECT Product,
       COUNT(*) AS Delivered_Orders,
       ROUND(SUM(TotalPrice), 2) AS Revenue_From_Delivered
FROM sales
WHERE OrderStatus = 'Delivered'
GROUP BY Product
ORDER BY Revenue_From_Delivered DESC
'''
run_query(query)
```

```python
# Q2: Products with average order value above 2000 (using HAVING)
query = '''
SELECT Product,
       ROUND(AVG(TotalPrice), 2) AS Avg_Order_Value,
       COUNT(*) AS Order_Count
FROM sales
GROUP BY Product
HAVING AVG(TotalPrice) > 2000
ORDER BY Avg_Order_Value DESC
'''
run_query(query)
```

```python
# Q3: Monthly revenue summary
query = '''
SELECT SUBSTR(Date, 1, 7) AS Year_Month,
       COUNT(*) AS Total_Orders,
       ROUND(SUM(TotalPrice), 2) AS Monthly_Revenue
FROM sales
GROUP BY SUBSTR(Date, 1, 7)
ORDER BY Year_Month ASC
'''
run_query(query)
```

```python
# Q4: Top 5 best revenue months
query = '''
SELECT SUBSTR(Date, 1, 7) AS Year_Month,
       ROUND(SUM(TotalPrice), 2) AS Monthly_Revenue
FROM sales
GROUP BY SUBSTR(Date, 1, 7)
ORDER BY Monthly_Revenue DESC
LIMIT 5
'''
run_query(query)
```

```python
# Q5: Cancellation rate per product
query = '''
SELECT Product,
       COUNT(*) AS Total_Orders,
       SUM(CASE WHEN OrderStatus = 'Cancelled' THEN 1 ELSE 0 END) AS Cancelled_Orders,
       ROUND(SUM(CASE WHEN OrderStatus = 'Cancelled' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS Cancel_Rate_Pct
FROM sales
GROUP BY Product
ORDER BY Cancel_Rate_Pct DESC
'''
run_query(query)
```

## Step 7 - Key Insights & Summary

```python
print('='*55)
print('      SQL ANALYSIS - KEY INSIGHTS')
print('='*55)

# Total orders and revenue
q1 = run_query('SELECT COUNT(*) as c, ROUND(SUM(TotalPrice),2) as r FROM sales')
print(f'\n1. DATASET OVERVIEW')
print(f'   Total Orders  : {q1["c"][0]}')
print(f'   Total Revenue : Rs. {q1["r"][0]:,.2f}')

# Top product by revenue
q2 = run_query('SELECT Product, ROUND(SUM(TotalPrice),2) as rev FROM sales GROUP BY Product ORDER BY rev DESC LIMIT 1')
print(f'\n2. TOP PRODUCT BY REVENUE')
print(f'   {q2["Product"][0]} with Rs. {q2["rev"][0]:,.2f}')

# Order status breakdown
q3 = run_query('SELECT OrderStatus, COUNT(*) as cnt FROM sales GROUP BY OrderStatus ORDER BY cnt DESC')
print(f'\n3. ORDER STATUS BREAKDOWN')
for _, row in q3.iterrows():
    print(f'   {row["OrderStatus"]:15} : {row["cnt"]}')

# Best payment method
q4 = run_query('SELECT PaymentMethod, COUNT(*) as cnt FROM sales GROUP BY PaymentMethod ORDER BY cnt DESC LIMIT 1')
print(f'\n4. MOST USED PAYMENT METHOD')
print(f'   {q4["PaymentMethod"][0]} with {q4["cnt"][0]} orders')

# Best month
q5 = run_query("SELECT SUBSTR(Date,1,7) AS ym, ROUND(SUM(TotalPrice),2) as rev FROM sales GROUP BY ym ORDER BY rev DESC LIMIT 1")
print(f'\n5. BEST REVENUE MONTH')
print(f'   {q5["ym"][0]} with Rs. {q5["rev"][0]:,.2f}')

# Top referral
q6 = run_query('SELECT ReferralSource, COUNT(*) as cnt FROM sales GROUP BY ReferralSource ORDER BY cnt DESC LIMIT 1')
print(f'\n6. TOP REFERRAL SOURCE')
print(f'   {q6["ReferralSource"][0]} brought the most orders: {q6["cnt"][0]}')

print('\n' + '='*5)
```
